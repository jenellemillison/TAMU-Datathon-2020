```python
import pandas as pd
unique_city_names = pd.read_csv("uniqueCities.csv", names=["City"])
cost_of_live = pd.read_csv("advisorsmith_cost_of_living_index.csv")
cost_of_live
```

```python
costs = cost_of_live.groupby(["City"]).mean()
cost_of_live = costs.filter(items=unique_city_names["City"], axis=0)
cost_of_live.to_csv("./processed_cost_of_live.csv")
```

```python
low = cost_of_live['Cost of Living Index'].quantile(0.0)
mid = cost_of_live['Cost of Living Index'].quantile(0.33)
high = cost_of_live['Cost of Living Index'].quantile(0.67)

lowdf = cost_of_live.loc[(cost_of_live['Cost of Living Index'] > low) & (cost_of_live['Cost of Living Index'] <= mid)]
middf = cost_of_live.loc[(cost_of_live['Cost of Living Index'] > mid) & (cost_of_live['Cost of Living Index'] <= high)]
highdf = cost_of_live.loc[(cost_of_live['Cost of Living Index'] > high)]

rating = []
for i in range(0, len(lowdf.index)):
    rating.append('Cheap')
lowdf["Cost of Living Rating"] = rating
rating.clear()

rating = []
for i in range(0, len(middf.index)):
    rating.append('Reasonable')
middf["Cost of Living Rating"] = rating
rating.clear()

rating = []
for i in range(0, len(highdf.index)):
    rating.append('Expensive')
highdf["Cost of Living Rating"] = rating
rating.clear()

all_ratings = pd.concat([lowdf, middf, highdf])
all_ratings
```

```python
rating_final = all_ratings.drop(['Cost of Living Index'], axis=1)
rating_final.to_csv(".\processed_cost_rating.csv")
```
