```python
import pandas as pd
unique_city_names = pd.read_csv("uniqueCities.csv", names=["City"])
quality_of_life = pd.read_csv("movehubqualityoflife.csv")
quality_of_life
```

```python
quality = quality_of_life.groupby(["City"]).mean()
quality_of_live = quality.filter(items=unique_city_names["City"], axis=0)
life_quality = quality_of_live["Quality of Life"]
life_quality.to_csv("./processed_quality_of_live.csv")
```

```python
low = quality_of_live['Quality of Life'].quantile(0.0)
mid = quality_of_live['Quality of Life'].quantile(0.33)
high = quality_of_live['Quality of Life'].quantile(0.67)

lowdf = quality_of_live.loc[(quality_of_live['Quality of Life'] > low) & (quality_of_live['Quality of Life'] <= mid)]
middf = quality_of_live.loc[(quality_of_live['Quality of Life'] > mid) & (quality_of_live['Quality of Life'] <= high)]
highdf = quality_of_live.loc[(quality_of_live['Quality of Life'] > high)]

rating = []
for i in range(0, len(lowdf.index)):
    rating.append('Life could be better')
lowdf["Quality of Life Rating"] = rating
rating.clear()



rating = []
for i in range(0, len(middf.index)):
    rating.append('Life is good')
middf["Quality of Life Rating"] = rating
rating.clear()

rating = []
for i in range(0, len(highdf.index)):
    rating.append('Doesn\'t get any better!')
highdf["Quality of Life Rating"] = rating
rating.clear()

all_ratings = pd.concat([lowdf, middf, highdf])
#all_ratings

lowdf
```

```python
rating_final = all_ratings.drop(['Quality of Life'], axis=1)
rating_final.to_csv(".\processed_quality_rating.csv")
```
