```python
import pandas as pd 
```

```python
df = pd.read_csv("ACSDP1Y2019.DP05_data_with_overlays_2020-10-17T174907.csv", header = 1)
```

```python
df1 = df[df.columns[[0,1,148,152,156,176,208,284]]]
```

```python
def GeoIdtoFIPS(string):
    """ Takes Census GEOID and converts to FIPS by taking last 5 characters """
    FIPS = string[-5:]
    return FIPS
```

```python
df1["FIPS"]=df1["id"].apply(GeoIdtoFIPS)
```

```python
df1.to_csv("Race_and_Ethnic_Data_County.csv")
```
